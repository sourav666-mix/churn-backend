```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv',usecols=[ 'gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'tenure', 'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'Contract', 'PaperlessBilling',
       'PaymentMethod', 'MonthlyCharges', 'TotalCharges', 'Churn'])
```

```python
def convert(text):
    if text=='Yes':
        return 1
    else:
        return 0

def convert2(obj):
  if obj=='No phone service':
     return 'No'
  else:
      return 'Yes'
```

```python
df['MultipleLines']=df['MultipleLines'].apply(convert2)
```

```python
cols_to_encode = ['MultipleLines','OnlineSecurity',  'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',

]

df[cols_to_encode] = df[cols_to_encode].apply(lambda col: col.map(convert2))
```

```python
columns_to_encode=[ 'MultipleLines','OnlineSecurity',  'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV','Churn','PaperlessBilling',
                   'Partner','Dependents','PhoneService'
  ]
df[columns_to_encode]  =df[columns_to_encode].apply(lambda col: col.map(convert))
```

```python
df['gender']=df['gender'].apply(lambda x: 1 if x=='Male' else 0)
```

```python
# Step 2 — encode it
df['InternetService'] = df['InternetService'].replace({
    'No'          : 0,
    'DSL'         : 1,
    'Fiber optic' : 2
}).astype(int)
```

```python
# Step 2 — encode it
df['Contract'] = df['Contract'].replace({
    'Two year'       : 0,
    'One year'        : 1,
    'Month-to-month' : 2
}).astype(int)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(inplace=True)
```

```python
# Step 2 — encode it
df['PaymentMethod'] = df['PaymentMethod'].replace({
    'Credit card (automatic)'         : 0,
    'Bank transfer (automatic)': 0,
    'Mailed check' : 1,
    'Electronic check':2
}).astype(int)
```

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace=True)
```

```python
x=df.iloc[:,:-1]
y=df['Churn']
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix, precision_score, classification_report
import numpy as np

# Train XGBoost model
xgb = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

# Cross-validation
cv = cross_val_score(xgb, x_train, y_train, cv=10, scoring='accuracy')
print("Max CV Accuracy:", np.round(np.max(cv), 2))
print("Mean CV Accuracy:", np.round(np.mean(cv), 2))

# Fit and predict
xgb.fit(x_train, y_train)
y_pred = xgb.predict(x_test)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

# Precision Score
precision = precision_score(y_test, y_pred, average='weighted')
print("\nPrecision Score:", np.round(precision, 2))

# Full report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# ==========================================
# IMPORTS
# ==========================================
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix, precision_score, classification_report
import numpy as np
import pickle
import os

# ==========================================
# 1. TRAIN XGBOOST MODEL
# ==========================================
xgb = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

# ==========================================
# 2. CROSS-VALIDATION
# ==========================================
cv = cross_val_score(xgb, x_train, y_train, cv=10, scoring='accuracy')
print("=" * 40)
print("        CROSS-VALIDATION RESULTS")
print("=" * 40)
print(f"  Max CV Accuracy  : {np.round(np.max(cv), 2)}")
print(f"  Mean CV Accuracy : {np.round(np.mean(cv), 2)}")
print(f"  Std CV Accuracy  : {np.round(np.std(cv), 2)}")

# ==========================================
# 3. FIT & PREDICT
# ==========================================
xgb.fit(x_train, y_train)
y_pred = xgb.predict(x_test)

# ==========================================
# 4. EVALUATION METRICS
# ==========================================
print("\n" + "=" * 40)
print("         CONFUSION MATRIX")
print("=" * 40)
cm = confusion_matrix(y_test, y_pred)
print(cm)

print("\n" + "=" * 40)
print("          PRECISION SCORE")
print("=" * 40)
precision = precision_score(y_test, y_pred, average='weighted')
print(f"  Precision : {np.round(precision, 2)}")

print("\n" + "=" * 40)
print("       CLASSIFICATION REPORT")
print("=" * 40)
print(classification_report(y_test, y_pred))

# ==========================================
# 5. SAVE & LOAD MODEL (Pickle)
# ==========================================
model_path = 'xgb_model.pkl'

# Save
with open(model_path, 'wb') as f:
    pickle.dump(xgb, f)
print(f"✅ Model saved to: {os.path.abspath(model_path)}")

# Load & Verify
with open(model_path, 'rb') as f:
    xgb_loaded = pickle.load(f)
print(f"✅ Model loaded successfully: {type(xgb_loaded)}")

# ==========================================
# 6. DOWNLOAD (Google Colab only)
# ==========================================
try:
    from google.colab import files
    files.download(model_path)
    print("✅ Download started!")
except ImportError:
    print(f"📁 Jupyter: Find your model at → {os.path.abspath(model_path)}")
```
