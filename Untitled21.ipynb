```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv',usecols=[ 'gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'tenure', 'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'Contract', 'PaperlessBilling',
       'PaymentMethod', 'MonthlyCharges', 'TotalCharges', 'Churn'])
```

```python
def convert(text):
    if text=='Yes':
        return 1
    else:
        return 0

def convert2(obj):
  if obj=='No phone service':
     return 'No'
  else:
      return 'Yes'
```

```python
df['MultipleLines']=df['MultipleLines'].apply(convert2)
```

```python
cols_to_encode = ['MultipleLines','OnlineSecurity',  'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',

]

df[cols_to_encode] = df[cols_to_encode].apply(lambda col: col.map(convert2))
```

```python
columns_to_encode=[ 'MultipleLines','OnlineSecurity',  'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV','Churn','PaperlessBilling',
                   'Partner','Dependents','PhoneService'
  ]
df[columns_to_encode]  =df[columns_to_encode].apply(lambda col: col.map(convert))
```

```python
df['gender']=df['gender'].apply(lambda x: 1 if x=='Male' else 0)
```

```python
# Step 2 — encode it
df['InternetService'] = df['InternetService'].replace({
    'No'          : 0,
    'DSL'         : 1,
    'Fiber optic' : 2
}).astype(int)
```

```python
# Step 2 — encode it
df['Contract'] = df['Contract'].replace({
    'Two year'       : 0,
    'One year'        : 1,
    'Month-to-month' : 2
}).astype(int)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(inplace=True)
```

```python
# Step 2 — encode it
df['PaymentMethod'] = df['PaymentMethod'].replace({
    'Credit card (automatic)'         : 0,
    'Bank transfer (automatic)': 0,
    'Mailed check' : 1,
    'Electronic check':2
}).astype(int)
```

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace=True)
```

```python
x=df.iloc[:,:-1]
y=df['Churn']
```

```python
import numpy as np

# First, ensure all columns in x_train and x_test are numeric.
# This step will convert any non-numeric strings (like ' ') to NaN.
x_train = x_train.apply(pd.to_numeric, errors='coerce')
x_test = x_test.apply(pd.to_numeric, errors='coerce')

# Now, identify and drop rows that contain NaN values introduced by the conversion.
# This ensures consistency between x_train and y_train, and x_test and y_test.
train_nan_rows_mask = x_train.isnull().any(axis=1)
if train_nan_rows_mask.any():
    print(f"Warning: Dropping {train_nan_rows_mask.sum()} rows from training set due to non-numeric values after conversion.")
    x_train = x_train[~train_nan_rows_mask]
    y_train = y_train[~train_nan_rows_mask]

test_nan_rows_mask = x_test.isnull().any(axis=1)
if test_nan_rows_mask.any():
    print(f"Warning: Dropping {test_nan_rows_mask.sum()} rows from test set due to non-numeric values after conversion.")
    x_test = x_test[~test_nan_rows_mask]
    y_test = y_test[~test_nan_rows_mask]

# Proceed with model fitting
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(precision_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
import pandas as pd

model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_split=5,
    class_weight='balanced',
    random_state=42
)

# Ensure all columns in x are numeric and handle any non-numeric values
x_cleaned = x.apply(pd.to_numeric, errors='coerce')

# Identify rows with NaN values introduced by conversion
nan_rows_mask = x_cleaned.isnull().any(axis=1)

# Drop rows with NaN values from both x_cleaned and y
if nan_rows_mask.any():
    print(f"Warning: Dropping {nan_rows_mask.sum()} rows from x and y for cross-validation due to non-numeric values after conversion.")
    x_cleaned = x_cleaned[~nan_rows_mask]
    y_cleaned = y[~nan_rows_mask]
else:
    y_cleaned = y

scores = cross_val_score(model, x_cleaned, y_cleaned, cv=10, scoring='accuracy')

print(f"F1 per fold : {scores.round(4)}")
print(f"Mean F1     : {scores.mean():.4f}")
print(f"Std F1      : {scores.std():.4f}")
```

```python
model_instance = RandomForestClassifier(random_state=42) # Create an instance of the classifier
cross=cross_val_score(model_instance,x_cleaned,y_cleaned,cv=10,scoring='accuracy') # Use the cleaned data from previous cell
print(np.round(np.mean(cross),2))
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

# Train and evaluate
kn = KNeighborsClassifier(n_neighbors=5, weights='distance', metric='manhattan')
kn.fit(x_train, y_train)
y_pred = kn.predict(x_test)

# Original metrics
print("Accuracy  :", accuracy_score(y_test, y_pred))
print("Precision :", precision_score(y_test, y_pred))
print("\nConfusion Matrix:\n",  confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Cross validation — 5 fold
cv_scores = cross_val_score(kn, x_cleaned, y_cleaned, cv=5, scoring='f1')

print("─" * 40)
print("Cross Validation Results (5-Fold)")
print("─" * 40)
print(f"F1 per fold : {cv_scores.round(4)}")
print(f"Mean F1     : {cv_scores.mean():.4f}")
print(f"Std F1      : {cv_scores.std():.4f}")
print(f"95% CI      : {cv_scores.mean():.4f} ± {cv_scores.std()*2:.4f}")
```

```python
models = {

    'RandomForest': {
        'model': RandomForestClassifier(random_state=42),
        'params': {
            'n_estimators'      : [100, 200, 300],
            'max_depth'         : [None, 5, 10, 20],
            'min_samples_split' : [2, 5, 10],
            'class_weight'      : ['balanced', None],
        }
    },
    'KNeighbors': {
        'model': KNeighborsClassifier(),
        'params': {
            'n_neighbors' : [3, 5, 7, 11, 15],
            'weights'     : ['uniform', 'distance'],
            'metric'      : ['euclidean', 'manhattan'],
        }
    },
     'DecisionTree': {
        'model': DecisionTreeClassifier(random_state=42),
        'params': {
            'max_depth'         : [None, 5, 10, 20],
            'min_samples_split' : [2, 5, 10],
            'criterion'         : ['gini', 'entropy'],
            'class_weight'      : ['balanced', None],
        }
    },

    'GaussianNB': {
        'model': GaussianNB(),          # replaced MultinomialNB
        'params': {
            'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5],
        }
    },
     'LogisticRegression': {
        'model': LogisticRegression(random_state=42, max_iter=1000),
        'params': {
            'C'            : [0.01, 0.1, 1, 10, 100],
            'penalty'      : ['l1', 'l2'],
            'solver'       : ['liblinear'],
            'class_weight' : ['balanced', None],
        }
    },

}
```

```python
results = {}

for name, config in models.items():
    print(f"Training {name}...")

    grid = GridSearchCV(
        estimator  = config['model'],
        param_grid = config['params'],
        cv         = 5,           # 5-fold cross validation
        scoring    = 'f1',        # f1 better than accuracy for imbalanced
        n_jobs     = -1,          # use all CPU cores
        verbose    = 0
    )
    grid.fit(x_train, y_train)

    best_model = grid.best_estimator_
    y_pred     = best_model.predict(x_test)
    acc        = accuracy_score(y_test, y_pred)

    results[name] = {
        'best_params' : grid.best_params_,
        'cv_f1_score' : round(grid.best_score_, 4),
        'test_accuracy': round(acc, 4),
        'model'       : best_model,
    }

    print(f"  best params : {grid.best_params_}")
    print(f"  cv f1 score : {grid.best_score_:.4f}")
    print(f"  test acc    : {acc:.4f}\n")
```
